Prediction of Online Shoppers Purchase Intentions.

1. Name the product most liked and purchased

2. Identify the special days when most products are purchased

3. List the top 5 special days of purchase and type of visitors.

4. Identify the months when most purchases happened

5. What the max and min revenue generated by the online shoppers

```python
import numpy as np
import pandas as pd
d = pd.read_csv('/content/online_shoppers_intention.csv')
d
```

```python
if 'Revenue' in d.columns and 'ProductRelated' in d.columns:
    purchased_products = d[d['Revenue'] == True]
    if not purchased_products.empty:
      # Find the product with the highest average 'ProductRelated' value among purchased items.
      product_engagement = purchased_products.groupby('ProductRelated')['Revenue'].count()
      most_liked_purchased = product_engagement.idxmax()
      print(f"The product most liked and purchased (based on ProductRelated interaction) is: {most_liked_purchased}")
    else:
        print("No purchases recorded in the dataset.")
else:
    print("The dataset does not contain 'Revenue' and/or 'ProductRelated' columns.")
```

```python
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'June', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
d['Month'] = pd.Categorical(d['Month'], categories=month_order, ordered=True)

d['Month_Visitor'] = d['Month'].astype(str) + '_' + d['VisitorType']

purchase_counts = d[d['Revenue'] == True].groupby('Month_Visitor')['Revenue'].count()

top_purchase_days = purchase_counts.sort_values(ascending=False).head(5)

# Display the results
print("Top 5 Special Days (Month_VisitorType) with the Most Purchases:")
top_purchase_days
```

```python

purchase_counts_by_month = d[d['Revenue'] == True].groupby('Month')['Revenue'].count()

# Find the month with the most purchases
month_with_most_purchases = purchase_counts_by_month.idxmax()

print(f"The month with the most purchases is: {month_with_most_purchases}")
```

```python

max_administrative = d['Administrative'].max()
min_administrative = d['Administrative'].min()

max_informational = d['Informational'].max()
min_informational = d['Informational'].min()

max_productrelated = d['ProductRelated'].max()
min_productrelated = d['ProductRelated'].min()


print(f"Maximum Administrative Duration: {max_administrative}")
print(f"Minimum Administrative Duration: {min_administrative}")
print(f"Maximum Informational Duration: {max_informational}")
print(f"Minimum Informational Duration: {min_informational}")
print(f"Maximum ProductRelated Duration: {max_productrelated}")
print(f"Minimum ProductRelated Duration: {min_productrelated}")
```

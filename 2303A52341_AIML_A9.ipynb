```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report
```

```python
df = pd.read_csv("/content/loan_data.csv")  # Replace with your file path
```

```python
print("Data Types:\n", df.dtypes)
```

```python
print("\nSummary Statistics:")
print(df.describe(percentiles=[.25, .5, .75, .9]))
```

```python
print("\nFeature Names:", df.columns.tolist())
```

```python
print("\nCount of Credit Approvals (credit.policy):")
print(df['credit.policy'].value_counts())
```

```python
sns.lmplot(x='fico', y='int.rate', hue='not.fully.paid', col='credit.policy', data=df, height=6, aspect=1)
plt.title("Trend of Interest Rate vs FICO by Credit Policy and Loan Status")
plt.show()
```

```python
df = pd.get_dummies(df, columns=['purpose'], drop_first=True)  # Dropping the first dummy to avoid multicollinearity
```

```python
X = df.drop(columns=['not.fully.paid'])  # Features
y = df['not.fully.paid']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
rf_model = RandomForestClassifier(random_state=42)
```

```python
rf_model.fit(X_train, y_train)
```

```python
y_pred = rf_model.predict(X_test)
```

```python
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
# Generate classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
nsimu = 21
accuracy=[0]*nsimu
ntree = [0]*nsimu
for i in range(1,nsimu):
    rfc = RandomForestClassifier(n_estimators=i*5,min_samples_split=10,max_depth=None,criterion='gini')
    rfc.fit(X_train, y_train)
    rfc_pred = rfc.predict(X_test)
    cm = confusion_matrix(y_test,rfc_pred)
    accuracy[i] = (cm[0,0]+cm[1,1])/cm.sum()
    ntree[i]=i*5
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x=ntree[1:nsimu],y=accuracy[1:nsimu],s=60,c='red')
plt.title("Number of trees in the Random Forest vs. prediction accuracy (criterion: 'gini')", fontsize=18)
plt.xlabel("Number of trees", fontsize=15)
plt.ylabel("Prediction accuracy from confusion matrix", fontsize=15)
```
